```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
#Low Pass Filter
fc = 200
N = 20
Fs = 1000
w_c = 2*fc/Fs                        #Normalized Frequency
t = signal.firwin(N,w_c,window='hamming')
[w,h] = signal.freqz(t,a=1)          #Frequency Response

w = Fs*w/(2*np.pi)                   #Conversion of radians into Hertz

h_db = 20*np.log10(abs(h))           #Conversion to decibal
plt.figure(figsize=(12,12))          #Figure size fixed to 12*12
plt.subplot(321)                     #Subplot number one
plt.plot(w,h_db)                     #Magnitude Response
plt.xlabel("Normalized Frequency")
plt.ylabel("Magnitude(db)")
plt.title("Frequency response")

plt.subplot(322)                     #Subplot number two
h_phase = np.unwrap(np.arctan2(np.imag(h),np.real(h)))
plt.plot(w,h_phase)
plt.xlabel('Noramlized Frequency')
plt.ylabel('Phase (radians)')
plt.title("Phase Response")
```
