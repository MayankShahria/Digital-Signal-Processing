```python
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
```

## IIR Filter Design

```python
## Low Pass Filter Design


Fs = 1000
n = 5
fc = 100                                 #cut-off frequency

normalized_freq = 2*fc/Fs                #normalized frequency
[b,a] = signal.butter(n,normalized_freq,btype='Lowpass') #butterworth filter by default Low Pass filter

[w,h] = signal.freqz(b,a,worN = 2000)    #Frequency Response
w = Fs*w/(2*np.pi)                       #Conversion of Frequency axis back into hertz

h_db = 20*np.log10(abs(h))               #Conversion into dB form

plt.figure(figsize=(12,12))
plt.subplot(321)
plt.plot(w, h_db)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude(dB)')

plt.subplot(322)
plt.plot(w, np.angle(h))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (radians)")
```

```python
## High Pass Filter

Fs = 1000
n = 5
fc = 100

normalized_freq = 2*fc/Fs
[b,a] = signal.butter(n,normalized_freq,btype='Highpass') #butterworth filter by default High Pass filter

[w,h] = signal.freqz(b,a,worN = 2000)                     #Frequency Response
w = Fs*w/(2*np.pi)                                        #Conversion of Frequency axis back into hertz

h_db = 20*np.log10(abs(h))                                #Conversion into dB form

plt.figure(figsize=(12,12))
plt.subplot(321)
plt.plot(w, h_db)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude(dB)')

plt.subplot(322)
plt.plot(w, np.angle(h))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (radians)")
```
